```python
import matplotlib.pyplot as plt
import numpy
```

```python
def fun(x):
  return -x * (numpy.exp(-x/2))
def fun1(x):
  return numpy.sin(numpy.pi*x) + 2*numpy.cos(2*numpy.pi*x)+3*numpy.sin(2*numpy.pi*x)*numpy.exp(-x/2)
def fun2(x):
  return 2*x*numpy.exp(-x)
```

```python
xvalues = numpy.linspace(0,10,10000)
yvalues = fun(xvalues)
yvalues1 = fun1(xvalues)
xvalues1 = numpy.linspace(0,10,10000)
yvalues2 = fun2(xvalues1)
plt.plot(xvalues,yvalues)
plt.plot(xvalues,yvalues1)
plt.plot(xvalues1,yvalues2)
fig, axs = plt.subplots(3, 1, figsize=(10, 12))
axs[0].plot(xvalues,yvalues)
axs[1].plot(xvalues,yvalues1)
axs[2].plot(xvalues1,yvalues2)
```

```python
def fun4(x,y):
  return numpy.sin(numpy.sqrt(2*x**2+y**2))
```

```python
x = numpy.linspace(-5, 5, 400)
y = numpy.linspace(-5, 5, 400)
X, Y = numpy.meshgrid(x, y)
Z= fun4(X,Y)
fig = plt.figure(figsize=(14, 7))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X,Y,Z,cmap='viridis', edgecolor='none')
ax2 = fig.add_subplot(122)
ax2.contour(X,Y,Z)
```

```python
def długość(f,a,b):
  x=np.linspace(a,b,1000)
  y=f(x)
  dx = np.diff(x)
  dy = np.diff(y)
  ds = np.sqrt(1 + (dy/dx)**2) * dx
  długość_krzywej = np.sum(ds)
  return długość_krzywej
print(długość(fun,0,10))
print(długość(fun1,0,10))
print(długość(fun2,0,10))
```
